```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

```python
test_new = pd.read_csv("test_new.csv",index_col=0)
test_new.head()
```

```python
train_new = pd.read_csv("train_new.csv", index_col = 0)
train_new.head()
```

```python
all_df = pd.concat([train_new, test_new], keys = ["train","test"])
all_df.head()
```

```python
all_dfnew = pd.get_dummies(all_df.loc[:, ["city","gender",'registered_via']])
```

```python
all_dfnew.head()
```

```python
all_get_df = pd.concat([all_df,all_dfnew],axis=1)
all_get_df.head()
```

```python
all_get_df["bd"].value_counts(ascending= True)
```

```python
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: int(x) if pd.notnull(x) else "NaN" )
```

```python
all_get_df[all_get_df["bd"]=="NaN"]
```

```python
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: 0 if (float(x)<0) else x)
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: 60 if float(x)>60 else x)
all_get_df["bd"].value_counts().head(100)
```

```python
bd = sorted(all_get_df["bd"].unique(), key = lambda x:float(x))
```

```python
sns.countplot(x="bd", data=all_get_df , order = bd)
plt.show()
```

# 如果0是0岁 0到20 标为10 20到40标为30 40以上标为50

```python
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: "0" if float(x)==0 else x)
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: "10" if(0<float(x)<20) else x)
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: "30" if (20<float(x)<40) else x)
all_get_df["bd"] = all_get_df["bd"].apply(lambda x: "50" if (40<float(x)) else x)
# for index, bd in enumerate(all_get_df["bd"]):
#     if float(bd) == 0:
#         all_get_df["bd"][index] = "0"
#     elif 0 < float(bd) <=20: 
#         all_get_df["bd"][index] = "baby"
#     elif 20 < float(bd) <=40:
#         all_get_df["bd"][index] = "teenager"
#     elif 40 <float(bd) :
#         all_get_df["bd"][index]= "Old"
                                        



```

```python
all_get_df["bd"].head(100)
```

```python
all_bd = pd.get_dummies(all_get_df["bd"],prefix="age")
all_get_df = pd.concat([all_get_df, all_bd],axis=1)
all_get_df.head()
```

```python
all_get_df.expiration_date = pd.to_datetime(all_get_df.expiration_date,format="%Y-%m-%d")
all_get_df.registration_init_time = pd.to_datetime(all_get_df.registration_init_time,format="%Y-%m-%d")
all_get_df.expiration_date = all_get_df.expiration_date.apply(lambda x: np.nan  if x=="NAN" else x)
all_get_df.registration_init_time = all_get_df.registration_init_time.apply(lambda x:np.nan  if x=="NAN" else x)
all_get_df['registration_duration'] = all_get_df.expiration_date - all_get_df.registration_init_time
all_get_df['registration_duration'] = all_get_df['registration_duration'] / np.timedelta64(1, 'D')
# all_get_df['registration_duration'] = all_get_df['registration_duration'].astype(int)

#---difference in months ---
all_get_df['registration_duration_M'] = (all_get_df.expiration_date - all_get_df.registration_init_time)/ np.timedelta64(1, 'M')
# all_get_df['registration_duration_M'] = round(all_get_df['registration_duration_M']).astype(int)
all_get_df.head(200)
```
