```python
import pandas as pd
from datetime import datetime
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
msno = train.drop('is_churn',1)
```

```python
tran = pd.read_csv("transactions.csv")
```

```python
new = pd.merge(left=msno, right=tran, how="left", on='msno')
```

```python
new.head()
```

```python
transation = new
transation.info()
```

```python
transation["payment_method_id"].isnull().value_counts()
```

```python
transation["payment_method_id"] = transation["payment_method_id"].apply(lambda x: int(x) if pd.notnull(x) else "NaN")
```

```python
transation = transation[transation["payment_method_id"] != "NaN"]
```

```python
transation.isnull().sum()
```

```python
transation["payment_plan_days"] = transation["payment_plan_days"].apply(lambda x: int(x))
transation["is_cancel"] = transation["is_cancel"].apply(lambda x: int(x))
transation["plan_list_price"] = transation["plan_list_price"].apply(lambda x: int(x))
transation["actual_amount_paid"] = transation["actual_amount_paid"].apply(lambda x: int(x))
transation["is_auto_renew"] = transation["is_auto_renew"].apply(lambda x: int(x))
```

```python
transation.head()
```

```python
transation["transaction_date"] = transation["transaction_date"].apply(lambda x:datetime.strptime(str(int(x)), '%Y%m%d').date())

transation["membership_expire_date"] = transation["membership_expire_date"].apply(lambda x:datetime.strptime(str(int(x)), '%Y%m%d').date())
```

```python
transation.head()
```

```python
transation["membership_expire_date"] = transation["membership_expire_date"].apply(lambda x:datetime.strptime(str(int(x)), '%Y%m%d').date())
```

```python
transation.head()
```

```python
transation.to_csv("transation_new.csv")
```
