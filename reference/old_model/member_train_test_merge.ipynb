# 手动更改

## 这里submiss是test data是train 从这里跑完改注释train test重新跑一遍。可得到train.csv



```python
all_f = submiss
# all_f = data
all_f
```

```python
all_ff=pd.merge(left=all_f,right=data1, on='msno',how='left')
# all_ff=pd.merge(all_ff,tran, on='msno',how='left')
all_ff.head()
```

```python
all_ff.info()
```

```python
all_ff["city"].isnull().value_counts()
```

```python
all_ff['city'] = all_ff.city.apply(lambda x: int(x) if pd.notnull(x) else "NAN")
all_ff["gender"] = all_ff.gender.fillna("NAN")
all_ff["registered_via"] = all_ff["registered_via"].apply(lambda x: int(x) if pd.notnull(x) else "NAN")
```

```python
all_ff.is_churn.isnull().value_counts()
```

```python
all_ff['registration_init_time'] = all_ff.registration_init_time.apply(lambda x: datetime.strptime(str(int(x)), "%Y%m%d").date() if pd.notnull(x) else "NAN" )
```

```python
all_ff["registration_init_time"].head()
```

```python
all_ff["expiration_date"] = all_ff["expiration_date"].apply(lambda x: datetime.strptime(str(int(x)),"%Y%m%d").date()if pd.notnull(x)
                                                            else"NAN")
all_ff["expiration_date"].head()
```
